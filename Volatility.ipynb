```python
from datetime import date, timedelta
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import statistics
```

```python
def volatility(data, str):
    data['Log_returns'] = np.log(data[str]/data[str].shift())
    volatility = data['Log_returns'][3:].std()*365**.5
    return volatility
```

```python
def log_returns(data,str):
    data['Log_returns'] = np.log(data[str]/data[str].shift())
    return  data['Log_returns']
```

```python
dataset_moderate = pd.read_csv('Moderate_investor.csv').copy()
dataset_optimal = pd.read_csv('Optimal_investor.csv').copy()
dataset_risky = pd.read_csv('Risky_investor.csv').copy()
dataset_conservative = pd.read_csv('Conservative_investor.csv').copy()
```

```python
dataset_moderate["bit"] = dataset_moderate["Bitcoin"]*dataset_moderate["Bitcoin_price"]
dataset_moderate["gld"] = dataset_moderate["Gold"]*dataset_moderate["Gold_price"]
dataset_moderate["combined"] = dataset_moderate["bit"]+dataset_moderate["gld"]

print(volatility(dataset_moderate,"bit"))
print(volatility(dataset_moderate,"gld"))
print(volatility(dataset_moderate,"combined"))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,4))
fig.tight_layout()
(data,str) = (dataset_moderate,'bit')
dataset = log_returns(data,'bit')
axes[0].hist(dataset[3:], width=0.05, color = "green")
axes[0].set_title(" Volatility curve: Bitcoin Assest Values")
axes[0].set_xlabel("Log returns")
axes[0].set_ylabel("Frequency of log returns")
axes[0].set_xlim((min(dataset[3:]), max(dataset[3:])))
axes[0].grid()

(data,str) = (dataset_moderate,'gld')
dataset = log_returns(data,'gld')
axes[1].hist(dataset[3:], width=0.01, color = "orange")
axes[1].set_title(" Volatility curve: Gold Assest Values")
axes[1].set_xlabel("Log returns")
axes[1].set_xlim((min(dataset[3:]), max(dataset[3:])))
axes[1].grid()

(data,str) = (dataset_moderate,'combined')
dataset = log_returns(data,'combined')
axes[2].hist(dataset[3:], width=0.05, color = "darkblue")
axes[2].set_title(" Volatility curve: Combined Assest Values")
axes[2].set_xlabel("Log returns")
axes[2].set_xlim((min(dataset[3:]), max(dataset[3:])))
axes[2].grid()

plt.savefig('volatility_moderate.png')
```
